```python
import camelot
import pandas as pd
tables = camelot.read_pdf('../dataset/suicide economic/2020 suicide economic.pdf', flavor="stream", pages="all")
```

```python
tables
```

```python
df0, df1 = tables[0].df, tables[1].df
```

```python
df0
```

```python
# Dropping first few columns
df0.drop([0,1,2,3,33,36,38,46], inplace=True)
# Renaming daman and diu
df0.loc[37, 0] = "31 D & N Haveli and Daman & Diu"

columns = ["State/UT", "less than `1 lakh - Male", "less than `1 lakh - Female", 
            "less than `1 lakh - Transgender", "less than `1 lakh - Total",
            "1 lakh & above - less than 5 lakhs - Male", "1 lakh & above - less than 5 lakhs - Female", 
            "1 lakh & above - less than 5 lakhs - Transgender", "1 lakh & above - less than 5 lakhs - Total",
            "5 lakhs & above - less than 10 lakhs - Male", "5 lakhs & above - less than 10 lakhs - Female", 
            "5 lakhs & above - less than 10 lakhs - Transgender", "5 lakhs & above - less than 10 lakhs - Total"]
df0.columns = columns 
df0["State/UT"] = df0["State/UT"].astype(str)
df0["State/UT"] = df0["State/UT"].str[2:].str.strip()

df0.reset_index(drop=True,inplace=True)

df0
```

```python
df1
```

```python
# Dropping first few columns
df1.drop([0,1,2,32,35,37,45], inplace=True)
# Dropping states row
df1.drop([0], axis=1, inplace=True)


columns = ["10 lakhs and above - Male", "10 lakhs and above - Female", 
            "10 lakhs and above - Transgender", "10 lakhs and above - Total",
            "Total - Male", "Total - Female", "Total - Transgender", "Total - Total",]
df1.columns = columns 
df1.reset_index(drop=True,inplace=True)
df1
```

```python
fin_df = pd.concat([df0, df1], axis=1)
```

```python
fin_df.to_csv("../dataset/suicide economic/2020 suicide economic.csv", index=False)
```
