```python
import camelot
import pandas as pd
tables = camelot.read_pdf('../dataset/suicide education/2020 suicide education.pdf', flavor="stream", pages="all")
```

```python
tables
```

```python
df0, df1, df2, df3 = tables[0].df, tables[1].df, tables[2].df, tables[3].df
```

```python
df0
```

```python
# Dropping first few columns
df0.drop([0,1,2,32,35,37,45], inplace=True)
# Renaming daman and diu
df0.loc[36, 0] = "31 D & N Haveli and Daman & Diu"



columns = ["State/UT", "No Education - Male", "No Education - Female", "No Education - Transgender", "No Education - Total",
            "Primary (up to class–5th) - Male", "Primary (up to class–5th) - Female", 
            "Primary (up to class–5th) - Transgender", "Primary (up to class–5th) - Total",
            "Middle (up to class–8th) - Male", "Middle (up to class–8th) - Female", 
            "Middle (up to class–8th) - Transgender", "Middle (up to class–8th) - Total"]
df0.columns = columns 
df0["State/UT"] = df0["State/UT"].astype(str)
df0["State/UT"] = df0["State/UT"].str[2:].str.strip()

df0.reset_index(drop=True,inplace=True)

df0
```

```python
df1
```

```python
# Dropping first few columns
df1.drop([0,1,2,32,35,37,45], inplace=True)
# Dropping states row
df1.drop([0], axis=1, inplace=True)


columns = ["Matriculate/Secondary (up to class–10th) - Male", "Matriculate/Secondary (up to class–10th) - Female", 
            "Matriculate/Secondary (up to class–10th) - Transgender", "Matriculate/Secondary (up to class–10th) - Total",
            "Hr. Secondary/Intermediate/Pre-University (up to class–12th) - Male", "Hr. Secondary/Intermediate/Pre-University (up to class–12th) - Female", 
            "Hr. Secondary/Intermediate/Pre-University (up to class–12th) - Transgender", "Hr. Secondary/Intermediate/Pre-University (up to class–12th) - Total",
            "Diploma/Certificate/ITI - Male", "Diploma/Certificate/ITI - Female", 
            "Diploma/Certificate/ITI - Transgender", "Diploma/Certificate/ITI - Total"]
df1.columns = columns 
df1.reset_index(drop=True,inplace=True)
df1
```

```python
df2
```

```python
# Dropping first few columns
df2.drop([0,1,2,32,35,37,45], inplace=True)
df2.drop([0], axis=1, inplace=True)

columns = ["Graduate and above - Male", "Graduate and above - Female", 
            "Graduate and above - Transgender", "Graduate and above - Total",
            "Professionals (MBA etc.) - Male", "Professionals (MBA etc.) - Female", 
            "Professionals (MBA etc.) - Transgender", "Professionals (MBA etc.) - Total",
            "Status Not known - Male", "Status Not known - Female", 
            "Status Not known - Transgender", "Status Not known - Total"]
df2.columns = columns 
df2.reset_index(drop=True,inplace=True)
df2
```

```python
df3
```

```python
# Dropping first few columns
df3.drop([0,1,2,32,35,37,45], inplace=True)
df3.drop([0,1,2],axis=1, inplace=True)

columns = ["Total - Male", "Total - Female", "Total - Transgender", "Total - Total"]
df3.columns = columns 
df3.reset_index(drop=True,inplace=True)
df3
```

```python
fin_df = pd.concat([df0, df1, df2, df3], axis=1)
```

```python
fin_df.to_csv("../dataset/suicide education/2020 suicide education.csv", index=False)
```
